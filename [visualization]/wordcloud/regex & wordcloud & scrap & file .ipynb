```python
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
import matplotlib as mpl

mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams["font.family"] = 'NanumGothic'
```

```python
import matplotlib.font_manager as font_manager

font_list = font_manager.findSystemFonts(fontpaths=None, fontext='ttf')

print('사용가능한 TTF 폰트 개수:', len(font_list))
font_list[:10] # 목록에서 처음 10 개만 보기
```

```python
# ready
txt=''
file=open(file='corona_finish.txt',encoding='utf-8-sig',mode='r')

while True:
    text=file.readline()
    if text=='':
        break
    txt+=text.strip()+' '
```

```python
fig,ax=plt.subplots(1,1,figsize=(6,4),dpi=200)

wordcloud=WordCloud(width=480, height=480,
                    margin=0,font_path='/usr/share/fonts/truetype/nanum/NanumSquareB.ttf').generate(txt)

ax.imshow(wordcloud, interpolation='bilinear')
ax.axis("off")
ax.margins(x=0, y=0)

plt.show()
```
