```python
data3['계약년월일']=data3['계약년월'].astype(str)+data3['계약일'].astype(str).apply(lambda day:day.zfill(2))
data3['계약년월일']=pd.to_datetime(data3['계약년월일'],format='%Y%m%d')

data3.drop(['계약년월','계약일'],axis=1,inplace=True)
# year
data3['계약연도']=data3['계약년월일'].dt.year

# diff
data3['건축_계약_차이']=data3['계약연도']-data3['건축년도']

# dong 
data3['동']=data3['시군구'].str.split(' ',expand=True)[1]
data3.drop('시군구',axis=1,inplace=True)

# exclude
# 누구나 아는 단지명이면 가격이 당연 높을거라고 생각하지만,가중치를 어떻게 처리할지 몰라 제거.
# 본번,부번 = 건물번호. 이용가치 x
data3.drop(['본번','부번','단지명'],axis=1,inplace=True)
```
