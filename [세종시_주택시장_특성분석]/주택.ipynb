```python
# iqr 방식을 이용한 아파트 매매 거래금액 이상치 비율.
# iqr 방식으로 봤을 때 2017~2020년 아파트 매매 거래금액의 크게 차이나는 것은 없다.

for x in data3['지역'].unique():
    temp=data3[data3['지역']==x]
    q_1=np.quantile(temp['거래금액(만원)'],0.25)
    q_3=np.quantile(temp['거래금액(만원)'],0.75)
    iqr=q_3-q_1
    value=np.mean(temp['거래금액(만원)'].apply(lambda x: q_1-iqr*1.5<x<q_3+iqr*q_3))
    color=Fore.RED if value < 0.6 else Fore.BLACK
    print(color+f'{x}의 {value:.2%}')
```

![](https://miro.medium.com/max/2974/1*fi4w1igkkPHzSBEv934FIw.png)

1sample t-test 목적 
  
관측된(observed) 혹은 수집된 샘플데이터를 통해 해당 샘플데이터가 모집단을 대표할 수 있을지에 대한 검정을 목적  
이때 t-test의 귀무가설은 ‘샘플데이터의 평균과 전체 데이터가 동일할 것이다’이다.  
`stats.ttest_1samp(mean,comp_mean)`

2sample t-test 목적  

2sample t-test는 샘플그룹 2개의 각 평균값이 서로 통계적으로 차이가 있는지 검정  
`stats.ttest_ind(s1,s1)`

chi-square (카이제곱 검정) 의 목적  

두 변수간의 연관도를 검정  
귀무가설은 기본적으로 두 변수가 ‘관계가 없다’로 시작
